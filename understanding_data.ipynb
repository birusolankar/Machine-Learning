```python
categorical_columns1 = ['Survived', 'Pclass', 'Sex', 'Embarked']
```

```python
for x in categorical_columns1:
  plt.figure()
  sns.countplot(data = titanic, x = x, palette = 'Set1')
  plt.xlabel(x)
  plt.ylabel('count')
```

```python
for x in categorical_columns1:
  columns_counts = titanic[x].value_counts()

  plt.figure()

  plt.pie(columns_counts, labels = columns_counts.index, autopct = '%1.1f%%', shadow=True, startangle=140)
  plt.title(f"distrubution of {x}")
  plt.axis('equal')
  plt.show()
```

```python
plt.pie(titanic['Pclass'].value_counts(), autopct = '%1.1f%%', shadow=True, startangle=140)
plt.show()
```
