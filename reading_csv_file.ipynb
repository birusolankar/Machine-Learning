```python
import matplotlib.pyplot as plt
```

```python
def candlestick(data, up_color='green', down_color='red'):
  diff = data['Close'] - data['Open']
  up = diff[diff>0]
  down = diff[diff<=0]

  fig = plt.figure(figsize=(8,6))
  for i in range(len(data)):
    if data.iloc[i]['Close']>data.iloc[i]['Open']:
      col = up_color
    else:
      col = down_color
    plt.vlines(x = i, ymin=data.iloc[i]['Low'], ymax=data.iloc[i]['High'], color=col, linewidth=2)
    plt.vlines(x = i, ymin=data.iloc[i]['Open'], ymax=data.iloc[i]['Close'], color=col, linewidth=6)

  labels = list(data.index.date)
  plt.xticks(ticks=np.arange(len(labels)), labels=labels, rotation=90)
  plt.ylabel('stock price')
  plt.title('price variation')
  plt.grid()
  plt.show()
```

```python
stock = reliance.head(50)
```

```python
candlestick(stock)
```
