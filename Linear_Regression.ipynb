```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/placement.csv")
```

```python
df.head(10)
```

```python
# plot data

plt.scatter(df['cgpa'], df['package'])
plt.xlabel('cgpa')
plt.ylabel('package(in lpa)')
```

```python
# preapring input and output column

X = df.iloc[:, 0:1]
y= df.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
# lets import train test split from sklear

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=2)
```

```python
X_train.shape
```

```python
# importing linear regression model from the sklearn
from sklearn.linear_model import LinearRegression
```

```python
# lets create an object of linera regression model
lr = LinearRegression()
```

```python
# now fit and train the model
lr.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
lr.predict(X_test.iloc[1].values.reshape(1,1))
```

```python
# lets plot the regression line

plt.scatter(df['cgpa'], df['package'])
plt.plot(X_test, lr.predict(X_test), color = 'red')
plt.xlabel('cgpa')
plt.ylabel('package(in lpa)')
```
