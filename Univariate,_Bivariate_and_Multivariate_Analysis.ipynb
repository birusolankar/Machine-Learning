```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
categorical_columns = ['Survived', 'Pclass', 'Embarked', 'Sex']
```

```python
for x in categorical_columns:
  plt.figure()
  sns.countplot(data=df, x= x, palette='Set1')
  plt.xlabel(x)
  plt.ylabel('Count')
  #plt.title('Embarked Count')
```

```python
for x in categorical_columns:
    column_counts = df[x].value_counts()

    plt.figure()
    plt.pie(column_counts, labels=column_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribution of {x}')
    plt.axis('equal')
    plt.show()
```
